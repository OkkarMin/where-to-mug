```python
from pprint import pprint
import pandas as pd
df = pd.read_json('./2021_1_data.json')
```

```python
def get_tut_details(index):
    result = []
    for i in index:
        details = i['details']
        for d in details:
            day = d['day']
            room = d['location']
            time = d['time']['full'].split('-')

            if d['type'] == 'TUT' and room != 'Online':
                result.append({'room': room, 'day': day, 'start_time': time})
    return result
```

```python
total_result = []
for course_code in df:
    total_result.extend(get_tut_details(df[course_code]['index']))
```

```python

def get_occupied_time_by_room(total_result):
    result = {}
    for item in total_result:
        room = item['room']
        occupied_day = item['day']
        occupied_time = item['time']

        # first time 'room' seen in the dict
        if not room in result.keys(): 
            result[room] = {occupied_day: [occupied_time]} # { 'TRX44': {'FRI': ['1030-1120'] }
        else:
            # first time 'occupied_day' was seen in the room
            if not occupied_day in result[room].keys():
                result[room][occupied_day] = [occupied_time]
            else:
                result[room][occupied_day].append(occupied_time) # 'TRX43': {'MON': ['0930-1020', '1030-1120', '1330-1420']
    
    return result
```

```python
occupied_by_room = get_occupied_time_by_room(total_result)
```

```python
import json
with open('room_occupancy.json', 'w') as outfile:
    json.dump(occupied_by_room, outfile, indent=2)
```
